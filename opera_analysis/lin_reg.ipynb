```python
all = [0.2973, 0.3283, 0.2615, 0.3665, 0.4471, 0.3395, 0.3453, 0.3834, 0.443, 0.3843, 0.3428, 0.5626, 0.3176, 0.5039, 0.5308, 0.5773, 0.3408, 0.2523, 0.2551]

positive = [0.4471, 0.3395, 0.3834, 0.443, 0.3843, 0.3428, 0.3176, 0.5039, 0.5308, 0.5773, 0.3408, 0.2511]
negative = list(set(all).symmetric_difference(set(positive)))

plt.plot(positive)
plt.plot(negative)
print(f'AVG POS: {np.mean(positive)} | AVG NEG: {np.mean(negative)}')
```

```python
def opera_to_df(input):
    # Import Raw
    raw_df = pd.read_csv(input, sep = '\t', skiprows = 9)
    raw_df = raw_df.dropna(axis = 'columns')
    return raw_df
```

```python
fields = opera_to_df('data/Objects_Population - coloc.txt')
fields
```

```python
all = list(fields['coloc - Spot Contrast'])
print(f'AVG HERE: {np.mean(all)}')

plt.plot(positive)
plt.plot(negative)
plt.plot(all[:100])
plt.axhline(y = 0.34)
```
